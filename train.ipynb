```python
import pandas as pd
```

```python
with open('image_paths.txt') as f:
    paths = [p.strip() for p in f.readlines()]
paths
```

```python
import numpy as np

y = np.load('targets.npy',allow_pickle=True)
```

```python
# Create a basic convolutional model using Keras

import tensorflow as tf
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Input, Reshape, Cropping2D

input_shape = (1200,1920)
```

```python
model = Sequential(
    [
        Input(shape=input_shape),
        Reshape((*input_shape, 1)),
        Cropping2D(cropping=((0, 0), (0, 1920-1200))),
        # Now immediately reduce the size with some large strides
        Conv2D(16, (3,3), strides=(2,2), activation='relu'),
        MaxPooling2D((2,2)),
        Conv2D(32, (3,3), activation='relu'),
        MaxPooling2D((2,2)),
        Conv2D(64, (3,3), activation='relu'),
        MaxPooling2D((2,2)),
        Conv2D(64, (3,3), activation='relu'),
        Flatten(),
        Dense(64, activation='relu'),
        Dense(14)
    ]    
)

model.summary()
```

```python
import keras
```

```python
class DataGenerator(keras.utils.Sequence):
    def __init__(self, paths, targets, batch_size=32, shuffle=True):
        super().__init__()
        self.paths = paths
        self.targets = targets
        self.batch_size = batch_size
        self.shuffle = shuffle
        self.on_epoch_end()
        
    def __len__(self):
        return int(np.floor(len(self.paths) / self.batch_size))
    
    def __getitem__(self, index):
        indexes = self.indexes[index*self.batch_size:(index+1)*self.batch_size]
        X, y = self.__data_generation(indexes)
        return X, y
    
    def on_epoch_end(self):
        self.indexes = np.arange(len(self.paths))
        if self.shuffle:
            np.random.shuffle(self.indexes)
            
    def __data_generation(self, indexes):
        X = np.empty((self.batch_size, *input_shape))
        y = np.empty((self.batch_size, 14))
        
        for i, index in enumerate(indexes):
            X[i] = keras.utils.load_img(self.paths[index], color_mode='grayscale')
            y[i] = self.targets[index]
            
        return X, y
    
```

```python
from sklearn.model_selection import train_test_split

train_paths, test_paths, train_targets, test_targets = train_test_split(paths, y, test_size=0.2)
test_paths, val_paths, test_targets, val_targets = train_test_split(test_paths, test_targets, test_size=0.5)

train_gen = DataGenerator(train_paths, train_targets)
val_gen = DataGenerator(val_paths, val_targets)
test_gen = DataGenerator(test_paths, test_targets)
```

```python
model.compile(optimizer='adam', loss='mean_absolute_error')
```

```python
model.fit(train_gen, validation_data=val_gen, epochs=10)
```
